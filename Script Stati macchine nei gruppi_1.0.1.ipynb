```python

import pandas as pd
import json
warnings.filterwarnings('ignore')

def filter_and_create_extradata(row, df_op):
    sdhMachineId = str(row['sdhMachineId'])
    start_time = pd.Timestamp(row['from'])
    end_time = pd.Timestamp(row['to'])

    filtered_df_op = df_op[
        (df_op['sdhMachineId'] == sdhMachineId) &
        (
            ((df_op['from_datetime'] <= end_time) & (df_op['from_datetime'] >= start_time)) | 
            ((df_op['to_datetime'] <= end_time) & (df_op['to_datetime'] >= start_time)) |
            ((df_op['from_datetime'] <= start_time) & (df_op['to_datetime'] >= end_time))
        )
    ]
    
    extradata = {}
    
    # Create extradata dictionary from filtered_df_op
    for _, op_row in filtered_df_op.iterrows():
        op_row_from = max(op_row['from_datetime'], start_time)
        op_row_to = min(op_row['to_datetime'], end_time)

        if op_row['correlationKey'] not in extradata:
            extradata[op_row['correlationKey']] = []
        
        extradata[op_row['correlationKey']].append({
            "properties_type": op_row["properties_type"],            
            "data_value": op_row["data_value"],
            "from": op_row_from.isoformat(),  # Convert to ISO format string
            "to": op_row_to.isoformat()       # Convert to ISO format string
        })
    
    return json.dumps(extradata)

df_status = pd.read_csv("./df_Status_.csv")
df_status_op = pd.read_csv("./df_Status_op_.csv")

df_status['from'] = pd.to_datetime(df_status['from'])
df_status['to'] = pd.to_datetime(df_status['to'])
df_status_op['from_datetime'] = pd.to_datetime(df_status_op['from_datetime'])
df_status_op['to_datetime'] = pd.to_datetime(df_status_op['to_datetime'])


# start_date = start_date.tz_localize(None)
# end_date = end_date.tz_localize(None)

if df_status_op['from_datetime'].dt.tz is not None :
    df_status_op['from_datetime'] = df_status_op['from_datetime'].dt.tz_localize(None)
else:
    pass

if df_status_op['to_datetime'].dt.tz is not None:
    df_status_op['to_datetime'] = df_status_op['to_datetime'].dt.tz_localize(None)
else:
    pass

df_status['extradata'] = df_status.apply(lambda row: filter_and_create_extradata(row, df_status_op), axis=1)

df_status["event_ID"] = df_status.index
df_status.set_index('event_ID')


df_status.to_csv("./df_Status_op_f.csv", index=False)

#>180m
```

```python
import pandas as pd
import json
import numpy as np
from datetime import datetime
from tqdm import tqdm

def print_with_timestamp(message):
    print(f"{datetime.now().isoformat()} - {message}")

def filter_and_create_extradata(row, filtered_df_op):
    sdhMachineId = str(row['sdhMachineId'])
    start_time = row['from']
    end_time = row['to']

    # Filtra ulteriormente per la macchina specifica
    machine_df_op = filtered_df_op[filtered_df_op['sdhMachineId'] == sdhMachineId]
    
    # Filtra per intervallo di tempo
    mask = (
        ((machine_df_op['from_datetime'] <= end_time) & (machine_df_op['from_datetime'] >= start_time)) | 
        ((machine_df_op['to_datetime'] <= end_time) & (machine_df_op['to_datetime'] >= start_time)) |
        ((machine_df_op['from_datetime'] <= start_time) & (machine_df_op['to_datetime'] >= end_time))
    )
    
    extradata = {}
    
    for _, op_row in machine_df_op[mask].iterrows():
        op_row_from = max(op_row['from_datetime'], start_time)
        op_row_to = min(op_row['to_datetime'], end_time)

        if op_row['correlationKey'] not in extradata:
            extradata[op_row['correlationKey']] = []
        
        extradata[op_row['correlationKey']].append({
            "properties_location": op_row["properties_location"],
            "properties_model": op_row["properties_model"],
            "properties_type": op_row["properties_type"],            
            "data_value": op_row["data_value"],
            "from": op_row_from.isoformat(),
            "to": op_row_to.isoformat()
        })
    
    return json.dumps(extradata)

print_with_timestamp("Loading CSV files...")
df_status = pd.read_csv("./df_Status_.csv")
# Specifica dtype come stringa per le colonne 2 e 3 per evitare il DtypeWarning
df_status_op = pd.read_csv("./df_Status_op_.csv", dtype={2: str, 3: str})

print_with_timestamp("Converting date columns to datetime...")
df_status['from'] = pd.to_datetime(df_status['from'])
df_status['to'] = pd.to_datetime(df_status['to'])
df_status_op['from_datetime'] = pd.to_datetime(df_status_op['from_datetime'])
df_status_op['to_datetime'] = pd.to_datetime(df_status_op['to_datetime'])

print_with_timestamp("Removing timezone from datetime columns if present...")
df_status_op['from_datetime'] = df_status_op['from_datetime'].dt.tz_localize(None)
df_status_op['to_datetime'] = df_status_op['to_datetime'].dt.tz_localize(None)

print_with_timestamp("Filtering df_status_op for relevant date range...")
min_date = df_status['from'].min()
max_date = df_status['to'].max()
filtered_df_op = df_status_op[(df_status_op['from_datetime'] <= max_date) & (df_status_op['to_datetime'] >= min_date)]

print_with_timestamp("Creating extradata for each row in df_status...")
tqdm.pandas()  # Inizializza tqdm con pandas
df_status['extradata'] = df_status.progress_apply(lambda row: filter_and_create_extradata(row, filtered_df_op), axis=1)

print_with_timestamp("Adding event_ID column and setting it as index...")
df_status["event_ID"] = df_status.index
df_status.set_index('event_ID', inplace=True)

output_file = "./df_Status_op_f.csv"
print_with_timestamp(f"Saving the result to {output_file}...")
df_status.to_csv(output_file, index=False)

print_with_timestamp("Process completed successfully.")
```
